```python
import numpy as np
import pandas as pd
Auto = pd.read_csv("Auto.csv")
Auto
```

```python
Auto_cleaned = Auto.dropna(subset=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration'])
Auto_cleaned
```

```python
Auto_cleaned['horsepower'] = pd.to_numeric(Auto_cleaned['horsepower'], errors='coerce')
Auto_cleaned = Auto_cleaned.dropna(subset=['horsepower'])
Auto_cleaned
```

```python
numpy_array=Auto_cleaned[['mpg','cylinders','displacement','horsepower','weight','acceleration']].values
numpy_array
```

```python
max_values_columnwise = np.max(numpy_array, axis=0)
max_values_columnwise
```

```python
min_values_columnwise = np.min(numpy_array, axis=0)
min_values_columnwise
```

```python
range=pd.DataFrame({'Minimum Value':min_values_columnwise,'Maximum value':max_values_columnwise}, 
                   index=['mpg','cylinders','displacement','horsepower','weight','acceleration'])
range
```

```python
mean=np.mean(numpy_array, axis=0)
mean
```

```python
std_dev=np.std(numpy_array, axis=0)
std_dev
```

```python
Qc=pd.DataFrame({'Mean':mean, 'Standard Deviation':std_dev}, index=['mpg','cylinders','displacement','horsepower','weight','acceleration'])
Qc
```

```python
numpy_array_rem=np.delete(numpy_array, np.s_[9:85], axis=0)
numpy_array_rem
```

```python
max_values_columnwise_rem = np.max(numpy_array_rem, axis=0)
max_values_columnwise_rem
```

```python
min_values_columnwise_rem = np.min(numpy_array_rem, axis=0)
min_values_columnwise_rem
```

```python
range=pd.DataFrame({'Minimum Value':min_values_columnwise_rem,'Maximum value':max_values_columnwise_rem}, index=['mpg','cylinders','displacement','horsepower','weight','acceleration'])
range
```

```python
mean_rem=np.mean(numpy_array_rem, axis=0)
mean_rem
```

```python
std_dev_rem=np.std(numpy_array_rem, axis=0)
std_dev_rem
```

```python
Qd=pd.DataFrame({'Mean':mean_rem, 'Standard Deviation':std_dev_rem}, index=['mpg','cylinders','displacement','horsepower','weight','acceleration'])
Qd
```
